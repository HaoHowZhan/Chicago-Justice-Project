```python
pdf = pd.DataFrame()
```

```python
path = './911call/other/'
files= os.listdir(path)
files
```

```python
for file in files:
    pd_temp = pd.read_csv(path+file,skiprows=1,header=None)
    pd_temp.columns = ['EVENTNUMBER', 'OCCCNT', 'CREATIONDATE', 'CREATIONDATEHN',
       'OVERRIDEDATE', 'OVERRIDEDATEHN', 'SEGMENTNAME', 'SEGMENTDATE',
       'SEGMENTDATEHN', 'CREATIONFLAG', 'SOURCEFLAG']
    print(pd_temp.shape)
    pdf = pd.concat([pdf,pd_temp])
    print(pdf.shape)
```

```python
pdf.head()
```

```python
pdf.CREATIONDATE = pd.to_datetime(pdf.CREATIONDATE, format='%m/%d/%Y %H.%M.%S',errors='coerce')
```

```python
pdf.OVERRIDEDATE = pd.to_datetime(pdf.OVERRIDEDATE, format='%m/%d/%Y %H.%M.%S',errors='coerce')
```

```python
pdf.SEGMENTDATE = pd.to_datetime(pdf.SEGMENTDATE, format='%m/%d/%Y %H.%M.%S',errors='coerce')
```

```python
pdf.to_csv('./911call/chifoia_event_segment_other.csv')
```
