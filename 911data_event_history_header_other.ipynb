```python
pdf = pd.DataFrame()
```

```python
path = './911call/other/'
files= os.listdir(path)
files
```

```python
for file in files:
    pd_temp = pd.read_csv(path+file,skiprows=1,header=None)
    pd_temp.columns = ['eventnumber', 'entrydate', 'ACKDATE', 'ADDROFOCCOVERRIDEFLAG',
       'ADVISEDFLAG', 'ALINUMBER', 'ARPFLAG', 'ATOM', 'ATOMPOL',
       'CALLERLOCSTREETDIR', 'CALLERLOCSTREETNAME', 'CANCELEDFLAG',
       'CASE_NUMBER_FLAG', 'CATCHUPFLAG', 'CLOSEDATE', 'CLOSEDATEHN',
       'CLRDATE', 'CONDISPGROUP', 'CONTACT', 'DISPATCHED_FLAG',
       'DISPDATE', 'findisposition', 'DISPWKS', 'DUPLICATE_FLAG',
       'ENRDATE', 'ENTRYDATEHN', 'ENTRYWKS', 'FILE_ADVISED_FLAG',
       'FINPRIORITY', 'FINRESPLEVEL', 'FINTYPE', 'INITDISPOSITION',
       'INITPRIORITY', 'INITRESPLEVEL', 'INITTYPE', 'LOCCITY',
       'LOCDISPGROUP', 'LOCOVERRIDEFLAG', 'LOCTYPEFLAG', 'OOSFLAG',
       'OTHERNOTIFIEDFLAG', 'RECDDATE', 'RECDDATEHN', 'REOPENFLAG',
       'RESPGROUP', 'RESPONSE', 'SOURCE', 'geo.ward', 'ali.answerdate',
       'ali.asqdate', 'ali.companyid', 'ali.direction', 'ali.equipment']
    print(pd_temp.shape)
    pdf = pd.concat([pdf,pd_temp])
    print(pdf.shape)
```
