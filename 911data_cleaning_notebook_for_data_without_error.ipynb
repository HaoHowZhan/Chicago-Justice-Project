```python
pdf = pd.DataFrame()
```

```python
path = './911call/chifoia_unit_segment/'
files= os.listdir(path)
files
```

```python
for file in files:
    pd_temp = pd.read_csv(path+file,sep="|",skiprows=1,header=None)
    pd_temp.columns = ['UNIT', 'OCCCNT', 'UNITLOGONDATE', 'CREATIONDATE',
       'CREATIONDATEHN', 'OVERRIDEDATE', 'OVERRIDEDATEHN', 'SEGMENTNAME',
       'SEGMENTDATE', 'SEGMENTDATEHN', 'WORKSTATION', 'PERSID', 'EVENTNUMBER',
       'CREATIONFLAG', 'SOURCEFLAG']
    print(pd_temp.shape)
    pdf = pd.concat([pdf,pd_temp])
    print(pdf.shape)
```
